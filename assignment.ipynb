# Week 1 Assignment: Basic Data Exploration and Cleaning using Pandas

### 1. Loading Dataset
 We'll be loading the 'combined_dataset.csv' (from Kaggle) into a Pandas DataFrame to begin our analysis.

```python
#importing Pandas;
import pandas as pd
#Loading the dataset;
df = pd.read_csv('Combined_dataset.csv')

print("Dataset loaded successfully!")
#verifying the dataset;
print(df.head())
```

### 2. Exploring the Data
Checking the shape, column names and data types  

```python
#getting dataset structure;
print("Dataset Structure (Shape) :", df.shape)
#getting the first 5 rows;
print("\nFirst 5 rows of the dataset :")
print(df.head())
#getting the last 5 rows;
print("\nLast 5 rows of the dataset :")
print(df.tail())
#getting the column information;
print("\nColumn information :")
print(df.info())
#getting data types explicitly as well(however) it is also shown in the info() method;
print("\nData types of each column :")
print(df.dtypes)
#getting summary statistics of the dataset;
df.describe()
```

### 3. Handling Missing Values
We'll first identify which columns contain null (NaN) values. And, then based on the count we'll decide whether to fill them with statistical averages or drop the rows to maintain the data quality.

```python
print("Initial Missing values:")
print(df.isnull().sum())
```

```python
#filling disciptive columns with a 'Not Available' value;
descriptive_columns = ['what_customers_said','seller_name','videos', 'seller_information','variations']
df[descriptive_columns] = df[descriptive_columns].fillna('Not Available')


#filling discount column with 0 as it is a numerical column and need to be filled with a numerical value;
df['discount'] = df['discount'].fillna(0)

# dropping remaining rows that might have critical missing values in other columns;
df.dropna(inplace=True)

#verifying the results after handling missing values;
print("Final Missing values count after handling :", df.isnull().sum().sum())

```

### 4. Performing Basic Operations (Filter & Select)
Demonstrating row filtering based on ratings and selecting a subset of relevant columns.

```python
# Selecting specific columns
selected_columns = df[['title', 'rating', 'final_price']]
print(selected_columns.head())

# Filter rows: Show only products with a rating higher than 4.0
high_rated_products = df[df['rating'] > 4.0]

print(f"Number of high-rated products: {len(high_rated_products)}")
high_rated_products.head()
```

### 5. Removing Duplicates
We'll ensure that each product entry is unique 

```python
# we will check the shape before removing duplicates to see how many duplicates were removed
print(f"Shape before removing duplicates: {df.shape}")
initial_rows = df.shape[0]
# now removing duplicate rows 
df.drop_duplicates(inplace=True)

final_rows = df.shape[0]

print(f"Shape after removing duplicates: {df.shape}")
print(f"Number of duplicate rows removed: {initial_rows - final_rows}")
```

### 6. Creating a Derived Column 'total_amount'
Before calculating the total amount, we must clean the price columns by removing currency symbols and commas, then convert them to numeric types. We will then calculate 'total_amount' as 'final_price' * 'quantity'.

```python

# We remove double quotes, currency symbol, and commas
df['final_price_cleaned'] = df['final_price'].str.replace('"', '', regex=False)
df['final_price_cleaned'] = df['final_price_cleaned'].str.replace('₹', '', regex=False)
df['final_price_cleaned'] = df['final_price_cleaned'].str.replace(',', '', regex=False)
df['final_price_cleaned'] = df['final_price_cleaned'].str.strip()

# Convert to numeric value, coercing errors to NaN (in case there are any non-numeric values left)
df['final_price_cleaned'] = pd.to_numeric(df['final_price_cleaned'], errors='coerce')

# Handle any remaining NaNs (just in case)
df['final_price_cleaned'] = df['final_price_cleaned'].fillna(0)

# let us now make a column 'quantity' and fill it with 1 as we are considering this quantity for each product;
df['quantity'] = 1

#now creating a derived 'total_amount'
df['total_amount'] = df['final_price_cleaned'] * df['quantity']

#verifying if our math works correctly by showing the relevant columns;
df[['title', 'final_price', 'final_price_cleaned', 'quantity', 'total_amount']].head()
```
