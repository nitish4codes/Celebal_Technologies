### 1. Loading Dataset
 We'll be loading the 'combined_dataset.csv' (from Kaggle) into a Pandas DataFrame to begin our analysis.

```python
#importing Pandas;
import pandas as pd
#Loading the dataset;
df = pd.read_csv('Combined_dataset.csv')

print("Dataset loaded successfully!")
#verifying the dataset;
print(df.head())
```

### 2. Exploring the Data
Checking the shape, column names and data types  

```python
#getting dataset structure;
print("Dataset Structure (Shape) :", df.shape)
#getting the first 5 rows;
print("\nFirst 5 rows of the dataset :")
print(df.head())
#getting the last 5 rows;
print("\nLast 5 rows of the dataset :")
print(df.tail())
#getting the column information;
print("\nColumn information :")
print(df.info())
#getting data types explicitly as well(however) it is also shown in the info() method;
print("\nData types of each column :")
print(df.dtypes)
#getting summary statistics of the dataset;
df.describe()
```

### 3. Handling Missing Values
We'll first identify which columns contain null (NaN) values. And, then based on the count we'll decide whether to fill them with statistical averages or drop the rows to maintain the data quality.

```python
print("Initial Missing values:")
print(df.isnull().sum())
```

```python
#filling disciptive columns with a 'Not Available' value;
descriptive_columns = ['what_customers_said','seller_name','videos', 'seller_information','variations']
df[descriptive_columns] = df[descriptive_columns].fillna('Not Available')


#filling discount column with 0 as it is a numerical column and need to be filled with a numerical value;
df['discount'] = df['discount'].fillna(0)

# dropping remaining rows that might have critical missing values in other columns;
df.dropna(inplace=True)

#verifying the results after handling missing values;
print("Final Missing values count after handling :", df.isnull().sum().sum())

```
